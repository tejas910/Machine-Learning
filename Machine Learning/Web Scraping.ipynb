```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://codewithharry.com"
r = requests.get(url)
htmlContent = r.content
print(htmlContent)
```

```python
soup = BeautifulSoup(htmlContent,'html.parser')
print(soup)
```

```python
title = soup.title
print(title)
```

```python
# Get Paragraph
paras = soup.find_all('p')
print(paras)
```

```python
# Get Anchor Tag
anc = soup.find_all('a')
all_links = set()
print(anc)
for link in anc:
    if(link.get('href') != '#'):
        linktext = "https://condewithharry.com"+link.get("href")
        all_links.add(linktext)
        print(linktext)
    
```

```python
print(soup.find('p')['class'])
```

```python
print(soup.find_all("p",class_="lead"))
```

```python
# Get The Text Of Tags
print(soup.find('p').get_text())
```
